```python
df = pd.read_csv(io.BytesIO(uploaded['EPA fuel economy dataset.csv']))
df = pd.DataFrame(df)
df
```

```python
#drop nonnumeric column from data frame
```

```python
non_floats = []
def dropNonnumeric(x):
    for col in x:
        if x[col].dtypes != "int":
           non_floats.append(col)
    
    x = x.drop(columns=non_floats)
    return(x)

df =dropNonnumeric(df)

print(df.head())
```

```python
#Create new column named "CityHighwayDiff" by differentiating "highway08" and "city08"
```

```python
CityHighwayDiff =  df['highway08'] -df['city08']
df['CityHighwayDiff'] = CityHighwayDiff
print(df.head())
```

```python
#plot a histogram of the column CityHighwaaydiff
```

```python
n, bins, patches = plt.hist(df['CityHighwayDiff'], density=True, edgecolor= "k", lw= 1, fc= "w")
plt.show
```

Question 2:

```python
# Using the EPA fuel economy data set. Consider the “year” column. 
#Create a new column “Decade” from the year column which maps each year into a class based on its decade.
```

```python
np.sort(df['year'])
decade = []
def yearTodecade(x):
    for year in x:
         decade.append((year // 10) * 10)
    return decade

df['decade'] = yearTodecade(df['year'])
df
```

```python
#Plot a histogram of the Decade column to visualize the occurrences in each of the decades.
```

```python
n, bins, patches = plt.hist(df["decade"],alpha=0.7, edgecolor= "k", lw= 1, fc= "w")
plt.show()
```

```python
#Remove any other non-numeric columns from the dataframe. 
#With the columns remaining, create a linear classifier that predicts the decade of a particular car based on the rest of the columns. 
#Train the classifier using the entire dataset. Calculate and print the training error of the resulting classifier.
```

```python
#preprocessing of the data
```

```python
from sklearn import preprocessing
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
X_train_minmax = min_max_scaler.fit_transform(df)
df = pd.DataFrame(X_train_minmax, columns = ['city08','cityA08','cylinders','highway08','highwayA08', 'id','year', 'CityHighwayDiff','decade'])

X = df[['city08','cityA08','cylinders','highway08','id','year','highwayA08','CityHighwayDiff']]

Y = df['decade']

print(X.head())
print(Y.head())

```

```python
#create a linear classifier that predicts the decade of a particular car based on the rest of the columns.
```

```python
from sklearn.linear_model import LogisticRegression 

from sklearn import metrics

from sklearn.model_selection import train_test_split 

from sklearn.metrics import mean_squared_error as mse

```

```python
#Train the classifier using the entire dataset. Calculate and print the training error of the resulting classifier.
```

```python
X_train, Y_train = X, Y
X_test = X
Y_test = Y
logreg = LogisticRegression()
model = logreg.fit(X_train, Y_train)
Y_predict = logreg.predict(X_test)
trainErr = mse(Y_train,Y_predict)

print("Shape of full dataset:", np.shape(X_train))
print("Error in Training, trained on full dataSet:" , trainErr)
```

```python
#Checking the accuracy and use the predictions find the confusion matrix
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix

from sklearn.metrics import classification_report
```

```python
confusion_matrix = confusion_matrix(Y_test, Y_predict)
print(confusion_matrix)
print("Accuracy:",metrics.accuracy_score(Y_test, Y_predict))
print(classification_report(Y_test, Y_predict))
plt.matshow(confusion_matrix)
plt.show()
```

```python
#Building on 3, split the training set used into a training and test set (split evenly 50:50).
#Re-fit the linear classifier developed above on the 50% training set.
#calculate and print the mean absolute error of the predictor using the 50% test set.
```

```python
#create a linear classifier ussing 50% of the data for traning and 50% for testing
#import mean absolute error
```

```python
from sklearn.metrics import mean_absolute_error as mae 

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5, random_state=42)

logreg = LogisticRegression()
model = logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
MAE = mae(Y_test,Y_pred)
print("Mean Absolute Error is:", MAE)

plt.scatter(Y_test, Y_pred, color = "blue")
plt.show()
```
